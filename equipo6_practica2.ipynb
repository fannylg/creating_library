```python
"""
Función para crear histogramas:

La función histogramas(df,varc) recibe dos argumentos: 'df' será el dataframe Cardiotocology y 'varc'
la lista que columnas que definimos como variables continuas. Se utiliza un ciclo FOR
para recorrer cada nombre de la lista varc, con ayuda de figure(figsize=(8, 6)) configuramos lo ancho y alto,
generemos el histograma con la variable en curso df[variable].hist configurando el color y dividiendo el rango de la variable en 20 bins,
por último agregamos títulos y etiquetas en los dos ejes.

"""

# Función para crear los histogramas para valiables continuas
def histogramas(df, varc):
    for variable in varc:
        plt.figure(figsize=(8, 6))
        df[variable].hist(color='skyblue', bins=20)
        plt.title(f'Histograma de {variable}')
        plt.xlabel(f'{variable}')
        plt.ylabel('Frecuencia')
        plt.grid(axis='y', alpha=0.75)
        plt.show()
histogramas(df, varc)
```

```python
"""
Función para crear boxplots:

La función boxplots(df,varc) recibe dos argumentos: 'df' será el dataframe Cardiotocology y 'varc'
la lista que columnas que definimos como variables continuas. Se utiliza un ciclo FOR
para recorrer cada nombre de la lista varc, con ayuda de figure(figsize=(8, 6)) configuramos lo ancho y alto,
generemos los boxplots con la paqueteria seaborn con la variable en curso sns.boxplot(x=df[variable] configuramos el color y
se agregan títulos y etiquetas en los ejes.

"""

# Función para crear los boxplots para valiables continuas
def boxplots(df, varc):
    for variable in varc:
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df[variable], color='skyblue')
        plt.title(f'Boxplot de {variable}')
        plt.xlabel('Valores')
        plt.ylabel(f'{variable}')
        plt.show()

boxplots(df, varc)
```

# Función Gráfico de barras
#### Se define una función grafico_barra_horizontal que nos permite visualizar la distribución de valores dentro de una variable en especifico de nuestro conjunto de datos utilizando los gráficos de barras con la librería Matplotlib. La función recibe dos argumentos en este caso nuestro df y variable, el nombre de la variable específica para la cual se desea generar el gráfico con plt.barh. Los argumentos pasados a plt.barh obtienen los valores únicos de la variable y su frecuencia. El código recorre una lista la cual definimos como nuestras variables descriptivas, la cual contiene las variables a analizar, para cada variable en vard, se llama a la función grafico_barra_horizontal pasandole el df creado y el nombre actual de la variables column como argumentos.

```python
def grafico_barra_horizontal(df, variable):
    plt.figure(figsize=(10, 6))
    plt.barh(df[variable].value_counts().index, df[variable].value_counts().values, color='skyblue')

    plt.title(f'Gráfico de Barras Horizontal de la variable {variable}')
    plt.xlabel('Frecuencia')
    plt.ylabel('Valores')

    plt.show()
```

```python
for column in vard:
    grafico_barra_horizontal(df, column)
```

# Gráficos de Líneas

### Se define la función 'graficos_lineas_series_temporales' la cual nos permite visualizar la evolución de las variables continuas a lo largo del tiempo utilizando la librería Matplotlib, la función recibe dos argumentos, en este caso nuestro df y 'variable', establecemos el tamaño de la figura utilizando plt.figure, se genera el gráfico de líneas utilizando plt.plot, los argumentos pasado a esta función, es la variables 'Date'  y la columna que contiene los valores de las variables, se establecen etiquetas tanto para el titulo como para el eje x que será con la etiqueta 'Fecha' y para el eje y será con la etiqueta 'Valor'. Se realiza una iteración sobre las variables de nuestra lista de variables continuas, se genera un gráfico

```python
def grafico_lineas_series_temporales(df, variable):
    plt.figure(figsize=(10, 6))
    plt.plot(df['Date'], df[variable])
    plt.xlabel('Fecha', fontsize=14)
    plt.ylabel('Valor', fontsize=14)
    plt.title(f'Gráfico de Líneas para Serie Temporal Unidimensional de {variable}', fontsize=16)
    plt.grid(True)
    plt.show()
```

```python
# Llamamos a la función para cada columna en varc
for column in varc:
    grafico_lineas_series_temporales(df, column)
```

## Funciones para crear heatmaps:

La primera función solicita dos parametros, el primero es el dataframe donde se encuentran los datos, y el segundo son las variables continuas de este dataframe, lo que realiza es hacer un dataframe donde se colocan las correlaciones que tienen estas variables continuas y posterior a ello las gráfica con la función heatmap de seaborn, con colores predeterminados y con la anotación igual a True.

Para la segunda función se solicita una variable que es el dataframe, sin embargo posterior a ello se piden por medio de inputs otros dos nombres, estos tienen que tener los nombres de la variable 1 y 2 que se quieren graficar y deben de ser variables discretas, posterior a ello se realiza un dataframe que es el resultado de la función crosstab, esta realiza una tabla dinamica para hacer el conteo de los cruces de los diferentes valores que toman estas variables, despues con ese dataframe resultante se utiliza la función de seaborn de heatmap, con colores predeterminados, anotación igual a True y con una variable que ayuda a como se visualiza esta anotación.



```python
def plot_heatmap(df,variables_continuas):
    df_corr=df[variables_continuas].corr()
    sns.heatmap(df_corr,cmap='RdYlGn',annot = True)
```

```python
plot_heatmap(df,varc)
```

## Funciones para crear graficos de densidad:

La función requiere de dos parametros, el primero es el dataframe donde se encuentran los datos, y el segundo puede ser una lista que contenga las variables continuas, posterior a ello se realiza un ciclo, esto con el fin de poder realizar la grafica para cada uno de los elementos que conforman la lista de las variables numericas, el ciclo toma la función de kdeplot de seaborn, la cual contiene como parametro de x la variable que se tiene en cuestión, se le rellena el area bajo la curva con el parametro de fill igual a true. Despues se coloca el titulo a cada una de las graficas y se muestra.


```python
def plot_densidad(df,varc):
  for col in varc:
    sns.kdeplot(x=df[col],fill=True)
    plt.title(f'Grafica de densidad de : {col}')
    plt.show()


```

```python
plot_densidad(df,varc)
```

Gráfica de violin

Esta función crea violinplots, que combinan las propiedades de los boxplots y las estimaciones de densidad de kernel. Los violinplots son útiles para visualizar la distribución de los datos y comparar múltiples variables.

```python
# Definimos la función para graficar violinplot
def plot_violine(df, varc, vard):
     for col_cont in vard:  # Iterar sobre cada variable continua
        for col_disc in varc:  # Iterar sobre cada variable discreta
            sns.set_style("whitegrid")
            plt.figure(figsize=(10, 6))  # Define el tamaño de la figura
            sns.violinplot(x=col_disc, y=col_cont, data=df, palette="muted")
            plt.title(f'Gráfico de Violín para {col_cont} vs {col_disc}')
            plt.xlabel(col_disc)  # Etiqueta del eje X
            plt.ylabel('Distribución de ' + col_cont)  # Etiqueta del eje Y
            plt.show()

plot_violine(df, varc, vard)
```
